```python
import pandas as pd
#Draw Box Plot
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_loc ="C:/Users/MD NADIM/Documents/NetBeansProjects/FindCochangeByClone/data_files/final_dataset/"
```

```python
subject_systems=["brlcad", "carol", "ctags", "freecol", "jabref", "jedit"]
clone_detectors=["conqat", "deckard", "iclones", "nicad5", "simcad", "simian"]
```

```python
for ss in subject_systems:
    df_results = pd.read_csv(data_loc+"result_"+ss+".csv", sep=',', header=0, encoding='utf-8')
    data=df_results.drop(['change_id','num_change'], axis=1)
    non_zero_data=[]
    #print(len(data))
    for c in clone_detectors:
        data_c=[]
        for i in range (0, len(data)):
            if(data[c][i]!=0):
                data_c.append(data[c][i])

        non_zero_data.append(data_c)  
        
    for c in range(6):
        sns.distplot(non_zero_data[c], hist = False, kde = True,
                 kde_kws = {'shade': False, 'linewidth': 2},
                 label = c)
        
    plt.legend(prop={'size': 10}, title = ss)
    plt.title('Density Plot')
    plt.xlabel('Delay (min)')
    plt.ylabel('Density')
    plt.show()
```

```python
fig= plt.figure()
plt.boxplot(non_zero_data, 0, '')
plt.xticks([1, 2, 3, 4, 5, 6], clone_detectors)
plt.grid(axis='x')
plt.xlabel("Machine Learning Algorithms")
plt.ylabel("F1 Score")
plt.show()
fig.savefig('./generated_img/cochange_box.png', dpi=300)
```
