```python
#Test data collection!!!
'''
38726 ['deckard', ' nicad5']
LUC:  ['176330', '280233']
LTC:  ['176334, 176344, 176329, 176337, 176328, 176348, 176339, 176335, 176345, 176340, 176336, 176346, 176331, 176332, 176338, 176342, 176330', '280226, 280224, 280234, 280232, 280229, 280225, 280227, 280230, 280228, 280223, 280233']
'''
# merged_tools=['deckard_2_0', ' nicad5']
# ltc=['176334, 176344, 176329, 176337, 176328, 176348, 176339, 176335, 176345, 176340, 176336, 176346, 176331, 176332, 176338, 176342, 176330', '280226, 280224, 280234, 280232, 280229, 280225, 280227, 280230, 280228, 280223, 280233']
def find_distinc_fragments(ss, tools, fragment_list):
    print("Merged Tools:", tools)
    print("Fragment List:", fragment_list)
    try:
        ########### Collect Fragments From Database....
        mySQLconnection = mysql.connector.connect(host='localhost',
                                 database='cochange_'+ss,
                                 user='root',
                                 password='')   
        cursor1 = mySQLconnection.cursor()
        records_arr=[]
        for i in range(len(tools)):
            print("i Index:", i)
            query="SELECT * FROM `clones_"+tools[i].strip()+"` WHERE `id` IN("+fragment_list[i].strip()+")"
            #print(query)
            cursor1.execute(query)
            records = cursor1.fetchall()
            records_arr.append(pd.DataFrame(records))
        collected_clones=pd.concat(records_arr).sort_values(by=[3, 4, 5])
        collected_clones=collected_clones.reset_index()
        ### --- 
        ########### Calculate Distinct Fragments in the Result....
        overlap=0
        fragments=1
        base=0
        for j in range(1, len(collected_clones)):  
            #print(base, i, overlap, fragments)
            #Dataframe refered by [col][row], 3=filename, 4=startline, 5=endline
            if(collected_clones[3][base]==collected_clones[3][j] and collected_clones[5][base]>=collected_clones[4][j]):
                overlap+=1
            else: 
                fragments+=1
                base=j

        #print(overlap, fragments)
        return fragments    
        ### --- 
    except Error as e:
            print("Error reading data from MySQL table", e, ss)
    finally:
        if (mySQLconnection.is_connected()):
            mySQLconnection.close()
            cursor1.close()
```

```python
# List of Subject Systems and Clone Detectors!!!
# subject_systems=["brlcad", "carol", "ctags", "freecol", "jabref", "jedit"]
subject_systems=["ctags"]
clone_detectors=["conqat", "deckard", "iclones", "nicad5", "simcad", "simian"]

# This is the declaration if you want to check the functinality for one subject sytems and clone detector. 
# subject_systems=["brlcad"]
# clone_detectors=["conqat"]

# Work for every subject systems. 
for ss in subject_systems:         
    # Read the cochange file stored in CSV. 
    df_luc_ltc = pd.read_csv(data_loc+"luc_ltc/"+ss+"_luc_ltc.csv", sep=',', header=0, encoding='utf-8')
    df_luc_ltc=df_luc_ltc.fillna(0).copy() # Fill the cell with no cochange by zero.

    #Make the header of the result file CSV
    #result_set = [["change_id", "num_used_clones", "num_total_clones"]]
    result_set = [["change_id"]]
    for tc in tools_combinations:
        if(tc!=''):
            result_set[0].append(tc.strip()+" (nuc)")
            result_set[0].append(tc.strip()+" (ntc)")
#     print(result_set)
        
    #Iterate each line of the csv file. 
    for i in range(len(df_luc_ltc)):
        line_accuracy=[df_luc_ltc["change_id"][i]] 
        
        #LUC and LTC are collected in Strings
        #for eache of the Merge Combinations. 
        for j in range(0, len(tools_combinations)): 
            # LUC: List of Used Clones, LTC: list of Total Clones. 
            luc=[] 
            ltc=[]
            combined_cochange=[]
            merged_tools=tools_combinations[j].split(',')
            luc_tools=[] #Tools having LUC values
            ltc_tools=[] #Tools having LTC values
            for mt in merged_tools:
                mt=mt.strip()
                col_name_luc=mt+"_luc"
                col_name_ltc=mt+"_ltc"
                luc_col_val=df_luc_ltc[col_name_luc][i]
                ltc_col_val=df_luc_ltc[col_name_ltc][i]
                #Take Only those tools and fragments which have values for both LUC and LTC. 
                if(luc_col_val!=0):
                    luc.append(luc_col_val.replace(" ", ""))
                    luc_tools.append(mt)
                    
                if(ltc_col_val!=0):
                    ltc.append(ltc_col_val.replace(" ", ""))
                    ltc_tools.append(mt)
                ### ---    
            print(df_luc_ltc["change_id"][i], merged_tools)
            print("LUC: ", luc)
            print("LTC: ", ltc)
            if(len(luc)>0):
                luc_count=find_distinc_fragments(ss, luc_tools, luc)
            else: luc_count=0
                
            if(len(ltc)>0):
                ltc_count=find_distinc_fragments(ss, ltc_tools, ltc)
            else: ltc_count=0
            print("LUC, LTC Counts:", luc_count, ltc_count)
            
            if(ltc_count>0):
                line_accuracy.append(luc_count/ltc_count)
        
        
        result_set.append(line_accuracy)
        if(i==10):
            break
            
    print(result_set)
#         


#     pd.DataFrame(result_set).to_csv(data_loc+"merged_recall_all/"+ss+"_merged_recall.csv", header=False, index=False)
#     print("Done: ", ss)
```

```python
#Count Distinct Fragments!!!
def remove_overlaps(collected_clones):
    overlap=0
    fragments=1
    base=0
    for i in range(1, len(collected_clones)):  
        #print(base, i, overlap, fragments)
        #Dataframe refered by [col][row], 3=filename, 4=startline, 5=endline
        if(collected_clones[3][base]==collected_clones[3][i] and collected_clones[5][base]>=collected_clones[4][i]):
            overlap+=1
        else: 
            fragments+=1
            base=i

    #print(overlap, fragments)
    return fragments
```
